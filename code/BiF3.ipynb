### Plot

```python
def plot_f(data, name, legend_size):
    plt.plot(data[:,0], data[:,1], label = name)
    plt.xlim(0, 700)
    plt.xlabel('Raman shift, cm^-1')
    plt.ylabel('Intensivity')
#     peaks = scipy.signal.find_peaks(data[:,1], distance = 1, height = 100, prominence = 40, width=7)
#     piki = []
#     for i in peaks[0]:
#         label =  r'$\lambda_i = $'
#         plt.axvline(x = data[:,0][i], color='red', label = label + str(round(data[:, 0][i], 1)) + " cm^-1", alpha = 0.5)
#         piki.append((data[:,0][i], data[:,1][i]))
#     plt.legend(prop={'size': legend_size})
    plt.grid()
#     print(piki)
def peaks(data, legend_size, prom=40, widht=7):
    peaks = scipy.signal.find_peaks(data[:,1], distance = 1, height = 400, prominence = prom, width=widht)
    piki = []
    for i in peaks[0]:
        label =  r'$\lambda_i = $'
        plt.axvline(x = data[:,0][i], color='red', label = label + str(round(data[:, 0][i], 1)) + " cm^-1", alpha = 0.5)
        piki.append((data[:,0][i], data[:,1][i]))
    plt.legend(prop={'size': legend_size})
    print(piki)
```

```python
def peaks(data, legend_size, prom=40, widht=7):
    peaks = scipy.signal.find_peaks(data[:,1], distance = 1, height = 400, prominence = prom, width=widht)
    piki = []
    for i in peaks[0]:
        label =  r'$\lambda_i = $'
        plt.axvline(x = data[:,0][i], color='red', label = label + str(round(data[:, 0][i], 1)) + " cm^-1", alpha = 0.5)
        piki.append((data[:,0][i], data[:,1][i]))
    plt.legend(prop={'size': legend_size})
    print(piki)
```
